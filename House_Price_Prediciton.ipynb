Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the California House Price Dataset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
print(house_price_dataset)
```

```python
#loading the data into a panadas dataframe
house_price_df = pd.DataFrame(house_price_dataset.data, columns= house_price_dataset.feature_names)
house_price_df.head()
```

```python
#add the target column to the dataframe
house_price_df['Price']=house_price_dataset.target
house_price_df.head()
```

```python
house_price_df.shape
```

```python
#check for missing values
house_price_df.isnull().sum()
```

```python
#statistical measures of all data
house_price_df.describe()
```

Understanding the correlation between various features in the dataset

```python
correlation = house_price_df.corr()
```

```python
#constructing a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8})
```

Splitting the data abd target

```python
X = house_price_df.drop(['Price'], axis=1)
Y = house_price_df['Price']
print(X)
print(Y)
```

Splitting the data into training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

Model training

```python
#loading the model
model=XGBRegressor()
```

```python
#trainig the model with x_train
model.fit(X_train,Y_train)
```

Evaluation

```python
#accuracy for predicitng on trianing data
training_data_prediction=model.predict(X_train)

print(training_data_prediction)
```

```python
#R squared Error
score_1 = metrics.r2_score(Y_train,training_data_prediction)
# mean absolute Erroe
score_2 = metrics.median_absolute_error(Y_train,training_data_prediction)
print("R squared error: ", score_1)
print("Mean absolute error: " ,score_2)
```

Visualizing the actual prices and predicitng prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predited prices")
plt.title("Actual prices vs Predicted prices")
plt.show()
```

Predicitng on test data

```python
#accuracy for predicitng on test data
test_data_prediction=model.predict(X_test)

print(test_data_prediction)
```

```python
#R squared Error
score_1 = metrics.r2_score(Y_test,test_data_prediction)
# mean absolute Erroe
score_2 = metrics.median_absolute_error(Y_test,test_data_prediction)
print("R squared error: ", score_1)
print("Mean absolute error: " ,score_2)
```
